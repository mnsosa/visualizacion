## Datasets

¿De dónde saco datos? Existen distintas fuentes, pero la principal es [Kaggle datasets](https://www.kaggle.com/datasets). 

También existe https://ourworldindata.org/, es interesante.

---

Por lo general, los datasets tienen origen en:
- **textos** (como ser comentarios en Google Play Store)
- **imágenes** (como ser las radiografías para predecir una enfermedad)
- **tablas** estructuradas (como ser una encuesta)

Las tablas pueden tener datos categóricos o datos numéricos.

Con **datos categóricos** se hace referencia a aquellos datos que pueden ser asignados en categorías. Esto va a depender del dato y el contexto.

## Ejemplo de uso

### Generación de un dataset 

La idea de generar un dataset con números arbitrarios es la de ir de a poco construyendo la idea de cómo funcionan las distintas librerías que se nombraron anteriormente. En la práctica, este paso será reemplazado por un probable `df = pd.read_csv(path_to_csv)`. 

````python
# Genero un conjunto del -50 al 50 de números enteros.
conjunto_1 = np.arange(-50, 50)

# Le aplico una función a cada dato
conjunto_2 = conjunto_1**2 

# Genero otro conjunto a partir del original
conjunto_3 = conjunto_1**3

# Genero otro conjunto 
conjunto_4 = conjunto_1**4
````

Aclaración. Miren cómo hubiera sido si en vez de usar `numpy` hubieramos usado listas:

```python
# Genero el primer array y lo convierto a lista
conjunto_1 = np.arange(-50, 50).tolist()

# Uso el list comprehension de python
conjunto_2 = [x**2 for x in conjunto_1] 
```



Veamos los datos que se generaron

````python
conjuntos = [conjunto_1, conjunto_2, conjunto_3, conjunto_4]

for i, c in enumerate(conjuntos):
  print(f"El conjunto {i} es:\n{c}\n")
````

### Matplotlib

Voy a decidir qué graficar y cómo.

Primero se me ocurre hacer un diagrama típico cartesiano, el de secundaria. Para ello usamos la función `plt.plot(x, y)`, dándole nosotros el x e y que queremos representar.

Hagamos **conjunto_1 vs conjunto_2**, siendo x, y respectivamente.

#### Básico

````python
# Defino el tamaño de la figura
plt.figure(figsize=(8, 8))

# Grafico conjunto_1 vs conjunto_2, pasándolo como x, y.
plt.plot(conjunto_1, conjunto_2)

# Muestro
plt.show()
````

#### Agregamos etiquetas y grid

````python
# Si no le digo de qué tamaño será la figura, lo hará con un tamaño
# definido por defecto.

# plt.figure(figsize=(12, 12))

# Le agrego el fondo cuadriculado
plt.grid(True)

# Ploteo de nuevo
plt.plot(conjunto_1, conjunto_2)

# Le agrego título
plt.title("Conjunto 1 vs Conjunto 2")
 
# Le agrego etiqueta al eje x
plt.xlabel("Conjunto 1")

# Le agrego etiqueta al eje y
plt.ylabel("Conjunto 2")

# Muestro
plt.show()
````

#### Definimos el tamaño de los ejes y ancho de línea

````python
plt.grid(True)

# Agregamos ancho de línea
plt.plot(conjunto_1, conjunto_2, linewidth=10)


plt.title("Conjunto 1 vs Conjunto 2")
plt.xlabel("Conjunto 1")
plt.ylabel("Conjunto 2")

# Agregamos el límite de los axis de esta manera
# [x_lim_inf, x_lim_sup, y_lim_inf, y_lim_sup]
plt.axis([-10, 10, 0, 200])

# Muestro
plt.show()

# Notar que es como si hubiéramos hecho un zoom
````

#### Varios datos en el mismo gráfico

````python
plt.figure(figsize=(10, 10))

plt.plot(conjunto_1, conjunto_2, 'ro',
         conjunto_1, conjunto_3, 'b^',
         conjunto_1, conjunto_4, 'g--')

plt.grid(True)
plt.show()
````

#### Fontdict

El fontdict es un diccionario de Python para dar estilo a las etiquetas del gráfico. A este diccionario lo usaremos como parámetro de las funciones de title, xlabel e ylabel.

````python
fontdict = {'family': 'monospace', 
            'color' : 'darkred',
            'weight': 'bold',
            'size'  : 18}
````

#### Agregando ylim, anotaciones y fontdict

````python
plt.figure(figsize=(10, 10))

plt.plot(conjunto_1, conjunto_2, 'ro',
         conjunto_1, conjunto_3, 'b^',
         conjunto_1, conjunto_4, 'd--')

# Otra forma de limitar los ejes (más fácil para mí)
# Notar que lo hago relativo al mínimo y máximo del conjunto 3, usando numpy
plt.ylim(np.min(conjunto_3), np.max(conjunto_3))

plt.title("Graficando todo al mismo tiempo", fontdict=fontdict)
plt.ylabel("Funciones", fontdict=fontdict)
plt.xlabel("Conjunto 1", fontdict=fontdict)

plt.annotate('Anotando el cero!', xy=(0, 0), xytext=(2, -35000),
             arrowprops=dict(facecolor='black', shrink=0.05),
             size=20)

plt.grid(True, color='darkblue', linestyle='--')
````

#### Subplots

````python
# Una sola figura para tres plots distintos
plt.figure(figsize=(18, 7))

# Indico que quiero un subplot de 1x3 y me paro en el 1 
plt.subplot(131)
plt.plot(conjunto_1, conjunto_2)
plt.title("C1 vs C2", fontdict=fontdict)
plt.ylabel("Conjunto 2", fontdict=fontdict)
plt.xlabel("Conjunto 1", fontdict=fontdict)
plt.grid(True)
plt.tight_layout() # Para que no se superpongan mucho

# Me paro en el 2 de este arreglo de plots 1x3
plt.subplot(132)
plt.plot(conjunto_1, conjunto_3)
plt.title("C1 vs C3", fontdict=fontdict)
plt.ylabel("Conjunto 3", fontdict=fontdict)
plt.xlabel("Conjunto 1", fontdict=fontdict)
plt.grid(True)
plt.tight_layout()

# Me paro en el 3 de este arreglo de plots 1x3
plt.subplot(133)
plt.plot(conjunto_1, conjunto_4)
plt.title("C1 vs C4", fontdict=fontdict)
plt.ylabel("Conjunto 4", fontdict=fontdict)
plt.xlabel("Conjunto 1", fontdict=fontdict)
plt.grid(True)
plt.tight_layout()

plt.show()
````

Cambiando la forma del subploteo. Ahora hago un arreglo de plots de 2x2, por lo que me entraría para 4 plots. Notar que yo solo tengo 3 plots, y queda un lugar vacío.

````python
# Una sola figura para tres plots distintos
plt.figure(figsize=(14, 8))

# Indico que quiero un subplot de 2x2 y me paro en el 1 
plt.subplot(221)
plt.plot(conjunto_1, conjunto_2)
plt.title("C1 vs C2", fontdict=fontdict)
plt.ylabel("Conjunto 2", fontdict=fontdict)
plt.xlabel("Conjunto 1", fontdict=fontdict)
plt.grid(True)
plt.tight_layout() # Para que no se superpongan mucho

# Me paro en el 2 de este arreglo de plots 2x2
plt.subplot(222)
plt.plot(conjunto_1, conjunto_3)
plt.title("C1 vs C3", fontdict=fontdict)
plt.ylabel("Conjunto 3", fontdict=fontdict)
plt.xlabel("Conjunto 1", fontdict=fontdict)
plt.grid(True)
plt.tight_layout()

# Me paro en el 3 de este arreglo de plots 2x2
plt.subplot(223)
plt.plot(conjunto_1, conjunto_4)
plt.title("C1 vs C4", fontdict=fontdict)
plt.ylabel("Conjunto 4", fontdict=fontdict)
plt.xlabel("Conjunto 1", fontdict=fontdict)
plt.grid(True)
plt.tight_layout()

plt.show()
````

#### Scatter

Es para hacer un gráfico punto a punto.

````python
plt.figure(figsize=(8, 8))
plt.grid(True)

# En vez de plot haceo scatter, que es graficar punto a punto
plt.scatter(conjunto_1, conjunto_2)

plt.title("Conjunto 1 vs Conjunto 2", fontdict=fontdict)
plt.xlabel("Conjunto 1", fontdict=fontdict)
plt.ylabel("Conjunto 2", fontdict=fontdict)
# Muestro
plt.show()
````

#### Ejemplo de gráfico 3D

````python
fig = plt.figure(figsize=(12, 12))

# Para que sea 3d (no es muy amistoso)
fig.add_subplot(projection='3d')

plt.plot(conjunto_1, conjunto_2, conjunto_3, 'r--')

plt.title("Graficando el conjunto 3 en función de los primeros 2", 
          position=(0.5, 1.05), 
          fontdict=fontdict)
plt.xlabel("Conjunto 1", size=14)
plt.ylabel("Conjunto 2", size=14)

plt.show()
````

````python
fig = plt.figure(figsize=(12, 12))

# Para que sea 3d (no es muy amistoso)
fig.add_subplot(projection='3d')

plt.plot(conjunto_1, conjunto_2, conjunto_4, 'b^')

plt.title("Graficando el conjunto 4 en función de los primeros 2", 
          position=(0.5, 1.05), 
          fontdict=fontdict)
plt.xlabel("Conjunto 1", size=14)
plt.ylabel("Conjunto 2", size=14)

plt.show()
````

### Pandas

#### Haciendo el primer DataFrame

A partir de los conjuntos generados en la sección anterior, procedo a armar un `dataframe` de pandas.

````python
# Genero una matriz con los conjuntos anteriores. Ignorar, no es de importancia.
data_conjuntos = np.stack(conjuntos, axis=1)

# Etiquetas de las columnas
column_labels = ["x", "f1", "f2", "f3"]

# Creo el objeto DataFrame de Pandas
df_conjuntos = pd.DataFrame(data=data_conjuntos, columns=column_labels)

df_conjuntos
````

````python
# Puedo cambiar el índice si quiero a que directamente sea x (el conjunto_1)
df_conjuntos.set_index("x", inplace=True)
df_conjuntos
````

Podemos ver la cantidad de días que incluye esto. Va desde enero del 2020 hasta marzo de 2022.

Analicemos únicamente los últimos 40 días.

````python
# Selecciono las últimas 40 filas (los últimos 40 días)
df_covid = df_covid.iloc[-40:-1]
df_covid
````

````python
df_covid.loc['3/14/22']
````

#### Separamos tres paises que queramos analizar

````python
argentina = df_covid["Argentina"]
spain     = df_covid["Spain"]
brazil    = df_covid["Brazil"]
````

````python
brazil
````

#### Comparemos visualmente

````python
plt.figure(figsize=(16, 12))

# Eje X son los índices
# Eje Y son los valores

plt.plot(brazil.index, brazil.values, 'go')
plt.plot(argentina.index, argentina.values, 'b^')
plt.plot(spain.index, spain.values, 'y')

# Agrego legend
plt.legend(("Brazil", "Argentina", "Spain"))

plt.title("Casos covid en los últimos 40 días", fontdict=fontdict)
plt.xlabel("Fechas", size=14)
plt.ylabel("Cantidad de casos", size=14)

plt.show()
````

````python
plt.figure(figsize=(16, 12))

# Se puede pasar directamente así, total ya contiene el par de (x, y)
plt.plot(brazil, 'go')
plt.plot(argentina, 'b^')
plt.plot(spain, 'y')

plt.legend(("Brazil", "Argentina", "Spain"))

plt.title("Casos covid en los últimos 40 días", fontdict=fontdict)
plt.xlabel("Fechas", size=14)
plt.ylabel("Cantidad de casos", size=14)

# Roto las etiquetas del eje X
plt.xticks(rotation=45)

plt.show()
````

````python
plt.figure(figsize=(16, 12))

# Divido los valores por 1 millón, así el eje queda en el orden de los millones
plt.plot(brazil / 1e6, 'g', linewidth=4)
plt.plot(argentina / 1e6, 'b', linewidth=4)
plt.plot(spain / 1e6, 'y', linewidth=4)

plt.legend(("Brazil", "Argentina", "Spain"))

plt.title("Casos covid en los últimos 40 días", fontdict=fontdict)
plt.xlabel("Fechas (mes/día/año)", fontdict=fontdict)
plt.ylabel("Cantidad de casos (millones)", fontdict=fontdict)

plt.xticks(rotation=45)

plt.show()
````

````python
plt.figure(figsize=(12, 8))

# Divido los valores por 1 millón, así el eje queda en el orden de los millones
plt.plot(argentina / 1e6, 'b', linewidth=4)
plt.plot(spain / 1e6, 'y', linewidth=4)

plt.legend(("Argentina", "Spain"))

plt.title("Casos covid en los últimos 40 días", fontdict=fontdict)
plt.xlabel("Fechas (mes/día/año)", fontdict=fontdict)
plt.ylabel("Cantidad de casos (millones)", fontdict=fontdict)

plt.xticks(rotation=45)

plt.show()
````

````python
plt.figure(figsize=(12, 8))

# Puedo hacer un scatter
plt.scatter(argentina.index, argentina.values)
plt.scatter(argentina.index, spain.values)

plt.legend(("Argentina", "Spain"))

plt.title("Casos covid en los últimos 40 días", fontdict=fontdict)
plt.xlabel("Fechas (mes/día/año)", fontdict=fontdict)
plt.ylabel("Cantidad de casos (millones)", fontdict=fontdict)

plt.xticks(rotation=45)

plt.show()
````

````python
plt.figure(figsize=(12, 8))

# Puedo hacer un gráfico de barra
plt.bar(argentina.index, argentina.values / 1e6)

plt.legend(("Argentina", "Spain"))

plt.title("Casos covid en los últimos 40 días", fontdict=fontdict)
plt.xlabel("Fechas (mes/día/año)", fontdict=fontdict)
plt.ylabel("Cantidad de casos (millones)", fontdict=fontdict)

plt.xticks(rotation=45)

plt.show()
````
